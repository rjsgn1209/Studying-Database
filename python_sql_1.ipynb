```python
# sakila 데이터베일스의 rantal 테이블에서 월별 렌탈수를 출력하시오..
sql = """
select * from rental;
"""

df = pd.read_sql(sql, db)
df['month'] = df['rental_date'].apply(lambda x: x.month_name())
df.groupby('month')['rental_id'].count()
```

```python
sql = """
select rental_id, left(rental_date, 7) as month, count(*) as 'rental_count' 
    from rental 
    group by month;
"""

df = pd.read_sql(sql, db)
df
```

```python
# 월별 총매출, 평균매출, 렌탈건수 합 을 출력하시오..
```

```python
sql = """
select * from payment;
"""

df_pay = pd.read_sql(sql, db)
df_pay
```

```python
sql = """
select * from rental;
"""

df_rental = pd.read_sql(sql, db)
df_rental
```

```python
df = pd.merge(df_pay, df_rental, on='rental_id')
df
```

```python
df['month'] = df.rental_date.dt.month
```

```python
df.groupby('month')['amount'].sum()
```

```python
# groupby 객체 메소드 : agg, filter, transform
agg: groupby 객체에 함수를 적용하는 메소드
fliter : 특정 group을 추출
transforms : 그룹의 값으로 데이터값을 바꿈
```

```python
sql = """
select count(payment_id) '구매건수', avg(amount) '평균구매액', sum(amount) '총구매액'\
    left(payment_date, 7) as month
    from payment
    group by month;
"""

payment_df = pd.read_sql(sql, db)
payment_df
```
