```python
# sakila database의 모든 table을 "테이블명.csv" 형식으로 folder sakila에 파일형식으로 저장하시오
```

```python
sql = """
show tables;
"""

cursor.execute(sql)
tables = cursor.fetchall()
```

```python
for i in tables:
    name = i[0]
    sql = 'select * from {}'.format(name)
    table = pd.read_sql(sql, db)
    table.to_csv("sakila/{}.csv".format(name))
```

```python
# 'GovernmentForm'에 'Republic'이라는 텍스트가 포함된 국가를 출력..
# (country name, city name)형식으로 묶어서 출력..
```
